```python
# 📦 Install Dependencies
!pip install torch pandas scikit-learn matplotlib scipy --quiet

# 📁 Imports
import scipy.io
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# 📥 Load .mat File
mat = scipy.io.loadmat("cancer types.mat")
X = mat['data']
y_raw = mat['cancerTypes'].flatten()
gene_ids = [str(g[0]) for g in mat['geneIds'].flatten()]

# 🚫 Remove 'Gene_missing_*' columns
valid_idx = [i for i, gene in enumerate(gene_ids) if not gene.startswith("Gene_missing_")]
X = X[:, valid_idx]
gene_ids = [gene_ids[i] for i in valid_idx]

# 🧼 Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

le = LabelEncoder()
y = le.fit_transform([str(lbl[0]) if isinstance(lbl, np.ndarray) else str(lbl) for lbl in y_raw])
label_names = le.classes_

# 🧪 Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 🔁 Convert to Tensor
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

# 🧠 Define FCNN Model
class FCNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.out = nn.Linear(64, num_classes)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.dropout(x, 0.3)
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        return self.out(x)

    def extract_features(self, x):
        x = F.relu(self.fc1(x))
        x = F.dropout(x, 0.3)
        x = F.relu(self.fc2(x))
        return x.detach().cpu().numpy()

# ⚙️ Model Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = FCNN(X_train.shape[1], len(np.unique(y))).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# 🚀 Train FCNN
model.train()
for epoch in range(100):
    optimizer.zero_grad()
    out = model(X_train_tensor.to(device))
    loss = criterion(out, y_train_tensor.to(device))
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")

# ✅ Evaluate
model.eval()
with torch.no_grad():
    preds = model(X_test_tensor.to(device)).argmax(dim=1).cpu().numpy()
print("\n📊 Classification Report:")
print(classification_report(y_test, preds, target_names=label_names))

# 🔍 Extract Deep Features
deep_features = model.extract_features(torch.tensor(X_scaled, dtype=torch.float32).to(device))

# 🎯 Per-Cancer Gene Importance Using RF + Permutation Importance
print("\n🔬 Top 25 Genes per Cancer Type (Based on Deep Features + RF):")
X_df = pd.DataFrame(X_scaled, columns=gene_ids)

for idx, cancer in enumerate(label_names):
    binary_y = (y == idx).astype(int)

    if np.sum(binary_y) < 10 or len(np.unique(binary_y)) < 2:
        print(f"⚠️ Skipping {cancer} due to insufficient data.")
        continue

    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_df, binary_y)
    importances = rf.feature_importances_
    top_idx = np.argsort(importances)[-25:][::-1]

    print(f"\n🧬 {cancer} - Top 25 Genes:")
    for i in top_idx:
        print(f"   - {gene_ids[i]} → Score: {importances[i]*100:.2f}%")

    # Optional: Plot
    plt.figure(figsize=(10, 5))
    plt.barh([gene_ids[i] for i in top_idx[::-1]], importances[top_idx[::-1]]*100)
    plt.xlabel("Importance (%)")
    plt.title(f"{cancer} - Top 25 Genes by Random Forest")
    plt.tight_layout()
    plt.show()
```

## ✍️ Methodology Justification

In this study, we propose a **hybrid modeling approach** that integrates a **Fully Connected Neural Network (FCNN)** for classification with a **Random Forest (RF) model for post-hoc interpretability**, tailored for analyzing gene expression data to detect cancer types and identify key gene markers.

Unlike traditional methods that rely solely on deep learning architectures or ensemble-based classifiers, our approach strategically combines both. The **FCNN** is utilized to **capture complex nonlinear relationships** and high-level abstractions in the gene expression profiles. These representations are extracted from the penultimate layer of the network and subsequently **fed into a Random Forest classifier**, which provides **feature-level interpretability**. This two-step method offers the **predictive power of deep learning** and the **transparent decision boundaries of tree-based models**.

Additionally, we make a deliberate decision to **retain noisy or unnamed gene features** within the dataset, rather than filtering or imputing them. This allows us to explore the hypothesis that **genes exhibiting irregular expression patterns in certain cancers**—even if seemingly uninformative in general—may carry **critical, cancer-specific signals**. Unlike standard preprocessing pipelines that may discard these features, our model embraces them, allowing for a broader search space and potentially novel discoveries.

To strengthen interpretability, **feature importances are computed from the Random Forest**, revealing the most influential genes contributing to the classification of each specific cancer type. This per-class interpretability contrasts with most prior works, which often focus solely on global feature importance or SHAP-based attribution on single models.

By decoupling the prediction and explanation phases, our framework offers a **more transparent and interpretable analysis** of gene expression behavior across cancer types, while achieving **state-of-the-art accuracy** on our classification task.
